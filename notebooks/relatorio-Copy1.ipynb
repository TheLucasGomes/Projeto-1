```python
import re

df_mrelevante = train.loc[train['target'] == 'muito relevante']

df_relevante = train.loc[train['target'] == 'relevante']

df_irrelevante = train.loc[train['target'] == 'irrelevante']


reviews_mrelevantes = df_relevante['review']

reviews_relevantes = df_relevante['review']

reviews_irrelevantes = df_irrelevante['review']


lista_reviews_mrelevantes = reviews_mrelevantes.to_list()

lista_reviews_relevantes = reviews_relevantes.to_list()

lista_reviews_irrelevantes = reviews_irrelevantes.to_list()




string_reviews_mrelevantes_raw = ''

for i in range(len(lista_reviews_mrelevantes)):

    string_reviews_mrelevantes_raw += str(lista_reviews_mrelevantes[i]) + ' '

string_reviews_relevantes_raw = ''

for i in range(len(lista_reviews_relevantes)):

    string_reviews_relevantes_raw += str(lista_reviews_relevantes[i]) + ' '

# se a def cleanup nao estiver funcionando pode ser q vc converteu tudo pra string
string_reviews_irrelevantes_raw = ''

for i in range(len(lista_reviews_irrelevantes)):

    string_reviews_irrelevantes_raw += str(lista_reviews_irrelevantes[i]) + ' '
    
    
    
def cleanup1(text):
    """
        Função de limpeza muito simples que troca alguns sinais básicos por espaços
    """
    #import string
    punctuation = '[´"!-/.:?;$''^\_`()\¨~%·¸=\\\]' # Note que os sinais [] são delimitadores de um conjunto.
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, '', text)
    return text_subbed

string_reviews_mrelevantes_semilimpa = cleanup1(string_reviews_mrelevantes_raw)

string_reviews_relevantes_semilimpa = cleanup1(string_reviews_relevantes_raw)

string_reviews_irrelevantes_semilimpa = cleanup1(string_reviews_irrelevantes_raw)


def cleanup2(text):
    """
        Função de limpeza muito simples que troca alguns sinais básicos por espaços
    """
    #import string
    punctuation = '[\n\t¯]' # Note que os sinais [] são delimitadores de um conjunto.
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, ' ', text)
    return text_subbed

string_reviews_mrelevantes_quaselimpa = cleanup2(string_reviews_mrelevantes_semilimpa)

string_reviews_relevantes_quaselimpa = cleanup2(string_reviews_relevantes_semilimpa)

string_reviews_irrelevantes_quaselimpa = cleanup2(string_reviews_irrelevantes_semilimpa)

lista_remover = ['ツ', '[ b]','[b]', '[ i]', '[i]', '[h1]', '[ h1]', '[*]', '[ list]','[list]', '[spoiler]', '[ spoiler]', '', '\t', 'ツ', '_', '+', '[strike]', '[ strike]',  '\\', '[]', '[url=https]', '[ url]', '[', ']',]




string_reviews_relevantes_limpa = ''

for i in lista_remover:
    if string_reviews_relevantes_limpa == '':
        string_reviews_relevantes_limpa = string_reviews_relevantes_quaselimpa 
    string_reviews_relevantes_limpa = string_reviews_relevantes_limpa.replace(i,'')

string_reviews_irrelevantes_limpa = ''

for j in lista_remover:
    if string_reviews_irrelevantes_limpa == '':
        string_reviews_irrelevantes_limpa = string_reviews_irrelevantes_quaselimpa 
    string_reviews_irrelevantes_limpa = string_reviews_irrelevantes_limpa.replace(j,'')
    
    
string_reviews_mrelevantes_limpa = '' 

for h in lista_remover:
    if string_reviews_mrelevantes_limpa == '':
        string_reviews_mrelevantes_limpa = string_reviews_mrelevantes_quaselimpa 
    string_reviews_mrelevantes_limpa = string_reviews_mrelevantes_limpa.replace(h,'')
    
    
    
    
    
string_reviews_mrelevantes_pronta = string_reviews_mrelevantes_limpa.lower() 
string_reviews_relevantes_pronta = string_reviews_relevantes_limpa.lower()
string_reviews_irrelevantes_pronta = string_reviews_irrelevantes_limpa.lower()

reviews_mrelevantes_analise = string_reviews_mrelevantes_pronta.split(' ')
reviews_relevantes_analise = string_reviews_relevantes_pronta.split(' ')
reviews_irrelevantes_analise = string_reviews_irrelevantes_pronta.split(' ')


serie_mrelevantes = pd.Series(reviews_mrelevantes_analise)
serie_relevantes = pd.Series(reviews_relevantes_analise)
serie_irrelevantes = pd.Series(reviews_irrelevantes_analise)

tabela_mrelevantes = serie_mrelevantes.value_counts()
tabela_relevantes = serie_relevantes.value_counts()
tabela_irrelevantes = serie_irrelevantes.value_counts()
    
    
```

```python
## Separar string grande e strings pequenas
## Suavização de Laplace

string_treinamento = string_reviews_mrelevantes_pronta+' '+string_reviews_relevantes_pronta + ' ' + string_reviews_irrelevantes_pronta


string_geral = string_treinamento 
lista_geral = string_geral.split(' ')

total_parcial = len(set(lista_geral))
total = total_parcial + len(lista_geral)
total
# (FREQUENCIA + 1)/(TOTAL + SET(TOTAL))

#reviews

#reviews
review = test.loc[:,'review'].to_list()


#listas absolutas
tabela_relevantes
tabela_irrelevantes

def classificador(review):
 
    lista_classificacao = []
    print(len(review))
    for i in review:   
        review = i

        review_semilimpa = cleanup1(review)
        review_quaselimpa = cleanup2(review_semilimpa)
        review_limpa = '' 
        for u in lista_remover:
            if review_limpa == '':
                review_limpa = review_quaselimpa
            review_limpa = review_limpa.replace(u,'')

        lista_review = review_limpa.split(' ')
        
        probabilidade_mrelevante = []
        probabilidade_relevante = []
        probabilidade_irrelevante = []

        
        for l in lista_review :
            try:
                probabilidade1 = (tabela_relevantes[l] + 1)/total
            except:
                probabilidade1 =  1/total  
            probabilidade_mrelevante.append(probabilidade1)
        
        
        for k in lista_review :
            try:
                probabilidade2 = (tabela_relevantes[k] + 1)/total
            except:
                probabilidade2 =  1/total  
            probabilidade_relevante.append(probabilidade2)
        for y in lista_review:
            try:
                probabilidade3 = (tabela_irrelevantes[y] + 1)/total
            except:
                probabilidade3 =  1/total
            probabilidade_irrelevante.append(probabilidade3)

        PM = np.prod(probabilidade_mrelevante)
        PR = np.prod(probabilidade_relevante)
        PI = np.prod(probabilidade_irrelevante)
        
        
        if PM > PR and PM > PI:
            lista_classificacao.append('muito relevante')
        elif PR> PI and PR>PM :
            lista_classificacao.append('relevante')
        elif PI > PR and PI > PM :
            lista_classificacao.append('irrelevante')
            
            
            
    return(lista_classificacao)

print('ACABOU') 

lista_final = classificador(review)

print(len(lista_final))
```

## Aperfeiçoamento:

Trabalhos que conseguirem pelo menos conceito B vão evoluir em conceito dependendo da quantidade de itens avançados:

* IMPLEMENTOU outras limpezas e transformações que não afetem a qualidade da informação contida nos tweets. Ex: stemming, lemmatization, stopwords

* Para Target com duas categorias: CRIOU pelo menos quatro categorias intermediárias de relevância baseadas na probabilidade: ex.: muito relevante, relevante, neutro, irrelevante, muito irrelevante
* EXPLICOU porquê não pode usar o próprio classificador para gerar mais amostras de treinamento
