### Montando um Classificador Naive-Bayes

Considerando apenas as mensagens da planilha Treinamento, ensine  seu classificador.

```python
import re

relev = train.loc[(train['target'] == 'relevante'),'review'].to_list()
mrelev = train.loc[(train['target'] == 'muito relevante'),'review'].to_list()
irrelev = train.loc[(train['target'] == 'irrelevante'),'review'].to_list()

review = train.loc[:,'review'].to_list()
review

relev2 = ''
mrelev2 = ''
irrelev2 = ''
for c in range(len(relev)):
    relev2 = relev2+' '+relev[c]
    
for c in range(len(mrelev)):
    mrelev2 = mrelev2+' '+mrelev[c]
    
for c in range(len(irrelev)):
    irrelev2 = irrelev2+' '+irrelev[c]


def cleanup(text):
    """
        Fun√ß√£o de limpeza muito simples que troca alguns sinais b√°sicos por espa√ßos
    """
    #import string
    punctuation = '[¬¥"!-.:?;$''{}()/]' # Note que os sinais [] s√£o delimitadores de um conjunto.
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern,'', text)
    return text_subbed



relev2 = relev2.split()
mrelev2 = mrelev2.split()
irrelev2 = irrelev2.split()


for c in range(len(relev2)):
    relev2[c] = cleanup(relev2[c])

for c in range(len(mrelev2)):
    mrelev2[c] = cleanup(mrelev2[c])
    
for c in range(len(irrelev2)):
    irrelev2[c] = cleanup(irrelev2[c])
    
    
    
relev3 = ''  
for i in range(len(relev2)):
    relev3 = relev3+' '+relev2[i]
pontuacao = "/\[x]‚òêüñæ''"

mrelev3 = ''  
for i in range(len(mrelev2)):
    mrelev3 = mrelev3+' '+mrelev2[i]
pontuacao = "/\[x]‚òêüñæ''"

irrelev3 = ''  
for i in range(len(irrelev2)):
    irrelev3 = irrelev3+' '+irrelev2[i]
pontuacao = "/\[x]‚òêüñæ''"





for c in pontuacao:
    relev3 = relev3.replace(c, " ")
    mrelev3 = mrelev3.replace(c, " ")
    irrelev3 = irrelev3.replace(c, " ")
    
print(mrelev3)
print(relev3)
print(irrelev3)
```

```python
## Separar string grande e strings pequenas
# (FREQUENCIA + 1)/(TOTAL + SET(TOTAL))

lista = []

lrelev = relev3.split()
lmrelev = mrelev3.split()
lirrelev = irrelev3.split()

serie_mrelev = pd.Series(lmrelev)
serie_relev = pd.Series(lrelev)
serie_irrelev = pd.Series(lirrelev)



avaliacoes = mrelev3+' '+relev3+' '+irrelev3
lavaliacoes = avaliacoes.split()

for z in range(len(lavaliacoes)):
    if lavaliacoes[z] == '' or lavaliacoes[z] == ' ':
        print(z)
        lavaliacoes.strip(lavaliacoes[z])
        



serie_aval = pd.Series(lavaliacoes)

tabela_aval = serie_aval.value_counts()
tabela_aval_r = serie_aval.value_counts(True)
tabela_mrelev = serie_mrelev.value_counts()
tabela_relev = serie_relev.value_counts()
tabela_irrelev = serie_irrelev.value_counts()

listatotal = avaliacoes.split(' ')
total = len(listatotal)
listaset = set(listatotal)


denominador = total+len(listaset)

def avaliador(review):
    for c in review:
        review = c
        
        review_sl = cleanup(review)
        review_l = ''  
        for i in range(len(review_sl)):
            review_l = review_l+' '+review_sl[i]
        for d in pontuacao:
            review_l = review_l.replace(d, " ")
            
        lwords = review_l.split(' ')
        
        prob_mrelevante = []
        prob_relevante = []
        prob_irrelevante = []

        for e in range(len(lwords)):
            word = lwords[e]
            try:
                prob = (tabela_mrelev[word] + 1)/denominador
            except:
                prob = 1/denominador
            prob_mrelevante.append(prob)

        for f in range(len(lwords)):
            word = lwords[f]
            try:
                prob = (tabela_relev[word] + 1)/denominador
            except:
                prob = 1/denominador
            prob_relevante.append(prob)
        for g in range(len(lwords)):
            word = lwords[g]
            try:
                prob = (tabela_irrelev[word] + 1)/denominador
            except:
                prob = 1/denominador
            prob_irrelevante.append(prob)




        PFM = np.prod(prob_relevante)
        PFR = np.prod(prob_mrelevante)
        PFI = np.prod(prob_irrelevante)
     
        listaprobs = [PFM,PFR,PFI]
        
        mprob = max(listaprobs)

        
        if mprob == PFM:
            ntarget = 'muito relevante'
            
        elif mprob == PFR:
            ntarget = 'relevante'
        
        else:
            ntarget = 'irrelevante'

        print (ntarget)
        print(mprob)
        print('---------------------------')
    
avaliador(review)
```
